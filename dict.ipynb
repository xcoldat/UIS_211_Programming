## Переменные

Переменная создаётся вписыванием названия переменной и её значения. От значения зависит тип переменной

Переменную нельзя назвать именем какой-либо команды

## Типы данных

integer (int) - целое число (..., -3, -2, -1, 0, 1, 2, 3, ...)

```python
a = 7
print(type(a))
```

string (str) - текстовая строка

```python
txt = "Строка"
print(type(txt))
```

float - нецелое число (0.0, 0.1, 0.11, ...)

```python
float_num = 2.3
print(type(float_num))
```

bool - булевая функция (true или false)

```python
boolean = True
boolean2 = False
print(type(boolean))
print(type(boolean2))
```

list - список - несколько значений в одном списке

```python
lst = [1, 2, 3]
print(type(lst))
```

dict - словарь - коллекция, где находятся пары значений. Первое - ключ, а второе - значение ключа.

```python
dct = {"key": "value"} # "key" - ключ, "value" - значение ключа
print(type(dct))
```

Можно изменить тип переменной, приписав перед названием переменной тип данных в следующей строке.

```python
a = 5 # создаётся переменная типа int и ей присваивается значение 5
print(float(a)) # тип данных меняется на float и выводится 5.0
```

## Условия

При помощи if() можно задавать условия и прописывать, что должно произойти в случае, если это условие соблюдается. Условие прописывается в скобочках. Через двоеточие прописываются дальнейшие действия.

```python
a = 1
if (a == 1):
    print("a = 1")
```

При помощи else можно задавать, что должно произойти, если ни одно условие не соблюдается.

```python
a = 2
if (a == 1):
    print("a = 1")
else:
    print("a != 1")
```

При помощи elif() можно задавать условия, которые будут проверяться, если не соблюдаются предыдущие условия.

```python
a = 2
if (a == 1):
    print("a = 1")
elif (a == 2):
    print("а = 2")
```

## Работа со строками и списками

len - вывод длины списка или строки в виде цифры типа int

```python
a = "hello"
print("Длина слова: " + str(len(a)))
```

переменная[i] - обозначение i-го пункта в строке/списке, где 0 - первый символ, 1 - второй, 2 - третий, -1 - последний, -2 - предпоследний и так далее.

```python
a = [1, 2, 3, 4]
print(a[0])
print(a[1])
print(a[-1])
print(a[-2])
```

append(a) - добавляет элемент a в конец списка

```python
lst = [1, 2, 3, 4]
lst.append(5)
print(lst)
```

insert(a, b) - добавляет элемент b на позицию с индексом a

```python
lst = [1, 3, 4]
lst.insert(1, 2)
print(lst)
```

extend([]) - добавляет несколько элементов в конец списка

```python
lst = [1, 2]
lst.extend([3, 4])
print(lst)
```

remove(a) - удаляет первый найденный элемент со значением a

```python
lst = [1, 2, 3, 4]
lst.remove(3)
print(lst)
```

pop()<br>
- если скобка пустая, то удаляет и возвращает последний элемент<br>
- если в скобке есть цифра, то удаляет элемент по индексу этой цифры

```python
lst = [1, 2, 3, 4]
lst.pop()
lst.pop(1)
print(lst)
```

del numbers[a] - удаляет число по индексу a

```python
lst = [1, 2, 3, 4, 5]
del lst[3]
print(lst)
```

clear() - удаляет все значения из списка

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]
lst.clear()
print(lst)
```

## Цикл for

Задаётся при помощи for a in b, где a - число, а b - название строкы/списка.<br>
При его помощи можно просматривать всю строку или список и выполнять определённые команды внутри их.

```python
string = 'string'
list = []
for i in string:
    list.append(i)

print(list) # Перенос букв из строки в список и вывод при помощи for
```

## Цикл while

При помощи while можно задать алгоритм, который будет выполняться, если выполняется условие. Если условие не выполняется, то цикл прекращает свою работу.

```python
i = 1
while i <= 5: # Вывод всех целых чисел от 1 до 5
    print(i)
    i += 1
```

В цикле while также работает break.

```python
i = 1
while True: # Вывод чисел от 1 до 5 с использованием break
    print(i)
    i += 1
    if i == 6:
        break
```

## Функции

def - команда для создания функций.<br>
После неё пишется название функции и скобочки, в них можно натсраивать, сколько переменных будет работать в функции.<br>
return возвращает переменную в конце функции<br>

```python
def hello(name):
    return name

my_name = input("Введите своё имя: ")
print("Привет, " + hello(my_name))
```

## Работа со словарями

dict.clear() - очищает словарь

```python
dct = {"key": "value"}
dct.clear()
print(dct)
```

dict.get(ключ) - возвращает значение указанного ключа

```python
dct = {"key": "value"}
print(dct.get("key"))
```

dict.items() - возвращает все пары

```python
dct = {"key": "value"}
print(dct.items())
```

dict.keys() - возвращает все ключи

```python
dct = {"key": "value"}
print(dct.keys())
```

dict.pop(ключ) - удаляет ключ и возвращает значение 

```python
dct = {"key": "value"}
print(dct.pop("key"))
print(dct)
```

dict.popitem() - удаляет и возвращает последнюю пару словаря

```python
dct = {"key": "value", "key2" : "value2"}
print(dct.popitem())
print(dct)
```
